```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
data = pd.read_csv('/content/insurance (1).csv')
ages_vector = data['age'].values
smoker_factor = pd.Categorical(data['smoker'])
data_list = [ages_vector, smoker_factor, data['expenses'].values]
matrix = np.array(data[['age', 'expenses']])
data_frame = data[['age', 'smoker', 'expenses']]
plt.figure(figsize=(10, 6))
plt.scatter(data['age'], data['smoker'].apply(lambda x: 1 if x == 'yes' else 0), alpha=0.6, c='blue', marker='+')
plt.title('Scatter Plot of Age vs Smoking Status', fontsize=14)
plt.xlabel('Age', fontsize=12)
plt.ylabel('Smoking Status (1: Yes, 0: No)', fontsize=12)
plt.grid(alpha=0.4)
plt.show()
filtered_data = data[(data['age'] >= 20) & (data['age'] <= 40)]
plt.figure(figsize=(10, 6))
plt.scatter(
    filtered_data['age'],
    filtered_data['smoker'].apply(lambda x: 1 if x == 'yes' else 0),
    alpha=0.6,
    c='green',
    marker='o',
)
plt.title('Scatter Plot of Age (20-40) vs Smoking Status', fontsize=14)
plt.xlabel('Age (20-40)', fontsize=12)
plt.ylabel('Smoking Status (1: Yes, 0: No)', fontsize=12)
plt.grid(alpha=0.4)
plt.show()
expenses_by_smoker = data.groupby('smoker')['expenses'].mean()
plt.figure(figsize=(8, 5))
expenses_by_smoker.plot(kind='bar', color=['orange', 'blue'], alpha=0.8, edgecolor='black')
plt.title('Average Medical Expenses of Smokers vs Non-Smokers', fontsize=14)
plt.ylabel('Average Expenses ($)', fontsize=12)
plt.xlabel('Smoking Status', fontsize=12)
plt.xticks(rotation=0)
plt.grid(axis='y', alpha=0.4)
plt.show()
smokers_by_gender = data[data['smoker'] == 'yes'].groupby('sex').size()
plt.figure(figsize=(8, 5))
smokers_by_gender.plot(kind='bar', color=['purple', 'cyan'], alpha=0.8, edgecolor='black')
plt.title('Number of Smokers by Gender', fontsize=14)
plt.ylabel('Count', fontsize=12)
plt.xlabel('Gender', fontsize=12)
plt.xticks(rotation=0)
plt.grid(axis='y', alpha=0.4)
plt.show()
plt.figure(figsize=(10, 6))
plt.scatter(data['age'], data['expenses'], alpha=0.6, c='red', marker='x')
plt.title('Scatter Plot of Age vs Medical Expenses', fontsize=14)
plt.xlabel('Age', fontsize=12)
plt.ylabel('Expenses ($)', fontsize=12)
plt.grid(alpha=0.4)
plt.show()
```
